```python
%matplotlib inline
from neuron import h
import matplotlib.pyplot as plt


# Create a section 
soma = h.Section()
soma.Ra = 1
soma.L = 100
soma.diam = 10
soma.cm = 1
```

```python
def transposeRecording(reclist, nloc, tstop):
    retlist = []  
    # each element of reclist is a neuron.h.Vector of recordings for one out of nloc spatial locations
    # (usually compartment centers) over time
    for i in range(0,int(tstop/h.dt)+1):
        retlist.append(h.Vector(nloc,0)) 
        for j in range(0,nloc):
            retlist[i].x[j] = reclist[j].x[i]
    return retlist
```

```python
locations = [0, 0.5, 1]
segs_lst = [3,7,21]

for location in locations:
    if location == 0:
        location_str = "beginning" 
    elif location == 0.5:
        location_str = "middle" 
    else:
        location_str = "end" 
        
    for nr_segs in segs_lst:
        # Inject a current step into the soma
        stim = h.IClamp(soma(location))         # Place a stimulation electrode in the middle of the soma
        stim.delay = 1 # ms
        stim.amp = 1   # pA
        stim.dur = 10  # ms


        # Initialize NEURON vectors to record time, voltage and current
        rec_t = h.Vector()
        rec_t.record(h._ref_t)

        rec_v = h.Vector()
        rec_v.record(soma(0.5)._ref_v)

        rec_i = h.Vector()
        rec_i.record(stim._ref_i)


        #Initialize and run a simulation
        h.load_file('stdrun.hoc')
        h.finitialize(-65)
        h.continuerun(5)
        h.steps_per_ms = 40.0

        soma.nseg = nr_segs

        reclist = h.List()

        for i in range(0,soma.nseg):
            tmpRecVec = h.Vector()
            tmpRecVec.record(soma((0.5 + i*1.0)/soma.nseg)._ref_v)
            reclist.append(tmpRecVec)

        # get the list of locations where recordings are happening
        xlocations = h.Vector()

        for i in range(0,soma.nseg):
            xlocations.append(soma.L*(0.5 + i*1.0)/(soma.nseg))

        h.init()
        h.run()

        # associate a voltage recording to each location in each time step
        spatialrec = transposeRecording(reclist, soma.nseg, 5)

        # plot the voltage in all the recording locations for a specific time step
        fig = plt.figure()
        plt.plot(xlocations, spatialrec[119]) # 2.975/0.025 = 119
        plt.ylabel("Voltage (mV)")
        plt.xlabel("Spatial location (um)")
        plt.suptitle("Location: {}, {} segments".format(location_str, nr_segs), y=1.05, fontsize=18)
        plt.title("Spatial distribution at t = 2.975 ms", fontsize=10)
```

The more segments we add the smoother the curve is. At each location a voltage is recorded at each time step. In the case of 3 segments, it is just connecting 3 points, while in the case of 21 segments it is connecting 21 points. By increasing the number of segments we are able to see better the change in voltage according to location.
Also, we can observe that the value of voltage decays exponentially as we are farther from the stimulation point.
In the case of stimulation in the beginning of the soma, the higher voltage is recorded in the stimulation point and then there is an exponential decay as we go more to the end of the soma. The same logic was observed in the other 2 location (middle and end).
